```python
import matplotlib.pyplot as plt 
import numpy as np
import math as mt
from mpl_toolkits import mplot3d
```

```python
#geometria del paralelepipedo
a = 5
b = 6
c = 3
# interaciones maximas de la serie de Fourier
nymMax = 25

#Avance de las variables Discretizadas x,y,z
divs = 10

#Matriz de resultados
ux1 = [] 

#Solucion de la Serie de Fourier
for x1 in range(divs + 1):
  x = a * (x1) / divs

  ux2 = [] 
  for x2 in range (divs + 1):
    y = b * (x2) / divs
  
    ux3 = []
    for x3 in range (divs + 1):
      z = c * (x3) / divs        
      T = 0
      for n in range (nymMax):
        for m in range (nymMax):
          beta = (n+1) * np.pi / a
          delta = (m+1) * np.pi / b
          alfa = np.sqrt(beta ** 2 + delta ** 2)
          R = (1 - mt.e ** (-2*alfa*c))
          Bprima = 4 * ((-1)**(n + 1) -1) * ((-1)**(m + 1) -1)* a * b / ((n + 1) * (m + 1) * (np.pi ** 2))
          Gx1 = np.sin(beta * x)
          Hx2 = np.sin(delta * y)
          Qx3 = -(mt.e ** (-2 * alfa * c)) * (mt.e ** (alfa * z)) + (mt.e ** (-alfa * z))
          T = T + Bprima * Gx1 * Hx2 * Qx3 / R
      print("u(", x,",", y,",", z,") =", T)     
      ux3.append(T)
    ux2.append(ux3) 
    print("\n")
  ux1.append(ux2)

```

```python
#Graficando los planos cada zi+1 = zi + salto 
for x3 in range(divs + 1):
  X = []          #Vectores a/salto de la variable x 
  Y = []          #Vectores b/salto de la variable y
  U = []          #Datos resultados de u(x,y,z)

  for x1 in range(divs + 1):
    #Arreglos bidimensionales para la grafica
    xtemp = []
    ytemp = []
    utemp=[]
    xtemp.append(a * (x1) / divs)
    xtemp = xtemp * (divs + 1)
    for x2 in range(divs + 1):
      ytemp.append(b * (x2) / divs)
      utemp.append(ux1[x1][x2][x3])
    U.append(utemp)
    X.append(xtemp)
    Y.append(ytemp)  
  #conviertiendo a los admitidos para a grafica 
  XX = np.array(X)
  YY = np.array(Y) 
  UU = np.array(U) 

  #Configuracion de la grafica
  fig = plt.figure(figsize = (14 , 9)) 
  ax = fig.add_subplot(111 , projection ='3d') 
  my_cmap = plt.get_cmap('hot')
  ax.plot_surface(XX , YY , UU , cmap = my_cmap) 
 
  ax.set_xlabel('x') 
  ax.set_ylabel('y') 
  ax.set_zlabel('u') 
  Tit = "Plano u = "+ str(x3 * c /divs)
  ax.set_title(Tit)
  plt.show()
  print("\n")
```
